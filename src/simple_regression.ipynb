# Linear Regression - TTNT

### Importing libraries

```python
import pandas as pd # Read data from file excel
import sklearn.linear_model as lm # Create model linear regression
import matplotlib.pyplot as plt # Draw graph
```

### Importing data

```python
df = pd.read_excel('../Data.xlsx')
df.head()
```

```python
plt.xlabel('Year')
plt.ylabel('Salary')

plt.scatter(df['Year'], df['Salary'], color='red', marker='.')
```

```python
x = df[['Year']]
y = df[['Salary']]

model = lm.LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)
```

### Spope and Intercept


```python
# Slope
model.coef_
```

```python
# Intercept
model.intercept_
```

```python
plt.plot(x, y_pred, color='blue', linewidth=2)
plt.scatter(df['Year'], df['Salary'], color='red', marker='.')
plt.show()
```
